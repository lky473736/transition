```python
import os
import math
import time
import random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
from torchmetrics import Accuracy, F1Score, Precision, Recall
from sklearn.metrics import silhouette_score, confusion_matrix
from sklearn.manifold import TSNE
import torchprofile
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display

save_plots = True

import os
import numpy as np
from sklearn.preprocessing import StandardScaler

def load_uci_har_raw(dataset_path):
    SIGNALS = ["body_acc_x", "body_acc_y", "body_acc_z", "body_gyro_x", "body_gyro_y", "body_gyro_z", "total_acc_x", "total_acc_y", "total_acc_z"]

    X_train_list, X_test_list = [], []
    for signal in SIGNALS:
        train_file = os.path.join(dataset_path, f"{signal}_train.txt")
        test_file = os.path.join(dataset_path, f"{signal}_test.txt")
        try:
            X_train_list.append(np.loadtxt(train_file, dtype=np.float32))
            X_test_list.append(np.loadtxt(test_file, dtype=np.float32))
        except Exception as e:
            print(f"Error loading file: {e}. Check dataset_path.")
            return None, None, None, None, None

    if not X_train_list: return None, None, None, None, None

    X_train = np.transpose(np.stack(X_train_list, axis=-1), (0, 2, 1))
    X_test = np.transpose(np.stack(X_test_list, axis=-1), (0, 2, 1))

    try:
        y_train = np.loadtxt(os.path.join(dataset_path, "y_train.txt"), dtype=int) - 1
        y_test = np.loadtxt(os.path.join(dataset_path, "y_test.txt"), dtype=int) - 1
    except Exception as e:
        print(f"Error loading label file: {e}")
        return None, None, None, None, None

    if np.isnan(X_train).any():
        train_nan_mask = ~np.isnan(X_train).any(axis=(1, 2))
        X_train = X_train[train_nan_mask]
        y_train = y_train[train_nan_mask]

    if np.isnan(X_test).any():
        test_nan_mask = ~np.isnan(X_test).any(axis=(1, 2))
        X_test = X_test[test_nan_mask]
        y_test = y_test[test_nan_mask]

    scaler = StandardScaler()
    num_samples_train, num_channels, sequence_length = X_train.shape
    num_samples_test = X_test.shape[0]

    X_train_reshaped = X_train.transpose(1, 0, 2).reshape(num_channels, -1).T
    X_test_reshaped = X_test.transpose(1, 0, 2).reshape(num_channels, -1).T

    scaler.fit(X_train_reshaped)
    X_train_scaled_reshaped = scaler.transform(X_train_reshaped)
    X_test_scaled_reshaped = scaler.transform(X_test_reshaped)

    X_train = X_train_scaled_reshaped.T.reshape(num_channels, num_samples_train, sequence_length).transpose(1, 0, 2)
    X_test = X_test_scaled_reshaped.T.reshape(num_channels, num_samples_test, sequence_length).transpose(1, 0, 2)

    ucihar_activity_names = ['Walking', 'Walking Upstairs', 'Walking Downstairs', 'Sitting', 'Standing', 'Laying']

    return X_train, y_train, X_test, y_test, ucihar_activity_names

class UCIHAR(Dataset):
    def __init__(self, X, y, normalize=True):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).long()
        if normalize:
            self.mu = self.X.mean(dim=(0,2), keepdim=True)
            self.sd = self.X.std(dim=(0,2), keepdim=True).clamp_min(1e-6)
        else:
            self.mu = self.sd = None

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        x = self.X[idx]
        if self.mu is not None:
            x = (x - self.mu.squeeze(0)) / self.sd.squeeze(0)
        return x, self.y[idx]

class ResidualBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv1 = nn.Conv1d(in_ch, out_ch, 3, padding=1)
        self.bn1 = nn.BatchNorm1d(out_ch)
        self.conv2 = nn.Conv1d(out_ch, out_ch, 3, padding=1)
        self.bn2 = nn.BatchNorm1d(out_ch)
        self.shortcut = nn.Conv1d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()

    def forward(self, x):
        res = self.shortcut(x)
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.bn2(self.conv2(x))
        return F.relu(x + res)

class MiniResNet(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.conv = nn.Conv1d(in_ch, 32, 3, padding=1)
        self.blocks = nn.Sequential(
            ResidualBlock(32, 64),
            ResidualBlock(64, d)
        )
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.conv(x)
        x = self.blocks(x)
        return self.pool(x)

class CNNLSTM(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(in_ch, 64, 5, padding=2),
            nn.ReLU(),
            nn.Conv1d(64, d, 5, padding=2),
            nn.ReLU()
        )
        self.lstm = nn.LSTM(d, d, batch_first=True)
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.cnn(x).transpose(1, 2)
        x, _ = self.lstm(x)
        return self.pool(x.transpose(1, 2))

class TCNBlock(nn.Module):
    def __init__(self, in_ch, out_ch, dilation):
        super().__init__()
        self.conv1 = nn.Conv1d(in_ch, out_ch, 3, padding=dilation, dilation=dilation)
        self.bn1 = nn.BatchNorm1d(out_ch)
        self.conv2 = nn.Conv1d(out_ch, out_ch, 3, padding=dilation, dilation=dilation)
        self.bn2 = nn.BatchNorm1d(out_ch)
        self.shortcut = nn.Conv1d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()

    def forward(self, x):
        res = self.shortcut(x)
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.bn2(self.conv2(x))
        return F.relu(x + res)

class TCN(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        dilations = [1, 2, 4, 8]
        ch = 64
        self.layers = nn.ModuleList([TCNBlock(in_ch if i==0 else ch, ch, dil) for i, dil in enumerate(dilations)])
        self.out = nn.Conv1d(ch, d, 1)
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        x = self.out(x)
        return self.pool(x)

class SeriesDecomp(nn.Module):
    def __init__(self, kernel_size):
        super().__init__()
        self.avg = nn.AvgPool1d(kernel_size, stride=1, padding=kernel_size//2)

    def forward(self, x):
        trend = self.avg(x)
        seasonal = x - trend
        return seasonal, trend

class AutoformerEncoderLayer(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.decomp = SeriesDecomp(25)
        self.attn = nn.MultiheadAttention(d, 4, batch_first=True)
        self.ff = nn.Linear(d, d)

    def forward(self, x):
        s, t = self.decomp(x)
        x = self.attn(x, x, x)[0] + x
        x = self.ff(x) + x
        return x

class AutoformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.proj = nn.Conv1d(in_ch, d, 1)
        self.layers = nn.ModuleList([AutoformerEncoderLayer(d) for _ in range(2)])
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.proj(x)
        for layer in self.layers:
            x = layer(x)
        return self.pool(x)

class PositionalEncoding(nn.Module):
    def __init__(self, d, max_len=128):
        super().__init__()
        pe = torch.zeros(max_len, d)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d, 2).float() * (-math.log(10000.0) / d))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]

class TransformerEncoderLayer(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.attn = nn.MultiheadAttention(d, 4, batch_first=True)
        self.ff = nn.Sequential(nn.Linear(d, d), nn.ReLU(), nn.Linear(d, d))
        self.norm1 = nn.LayerNorm(d)
        self.norm2 = nn.LayerNorm(d)

    def forward(self, x):
        x2 = self.norm1(x)
        x = x + self.attn(x2, x2, x2)[0]
        x2 = self.norm2(x)
        x = x + self.ff(x2)
        return x

class TransformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.proj = nn.Linear(in_ch, d)
        self.pe = PositionalEncoding(d)
        self.layers = nn.ModuleList([TransformerEncoderLayer(d) for _ in range(2)])
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = x.transpose(1, 2)
        x = self.proj(x)
        x = self.pe(x)
        for layer in self.layers:
            x = layer(x)
        return self.pool(x.transpose(1, 2))

class TSTransformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.proj = nn.Conv1d(in_ch, d, 1)
        self.pe = PositionalEncoding(d, 128)
        self.layers = nn.ModuleList([TransformerEncoderLayer(d) for _ in range(2)])
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.proj(x).transpose(1, 2)
        x = self.pe(x)
        for layer in self.layers:
            x = layer(x)
        return self.pool(x.transpose(1, 2))

class PyramidAttention(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.attn = nn.MultiheadAttention(d, 4, batch_first=True)

    def forward(self, x):
        # --- 기존 코드 ---
        # levels = [x[:, ::2**i] for i in range(3)]
        # attns = [self.attn(l, l, l)[0] for l in levels]
        # return sum(attns) / len(attns)

        # --- 수정된 코드 ---
        levels = [x[:, ::2**i] for i in range(3)]
        attns = [self.attn(l, l, l)[0] for l in levels]

        # 가장 긴 시퀀스 길이를 기준으로 설정
        target_len = attns[0].size(1)

        upsampled_attns = []
        for attn_out in attns:
            # 현재 텐서의 길이가 target_len보다 작으면 업샘플링 수행
            if attn_out.size(1) < target_len:
                # F.interpolate는 (B, C, L) 형태를 요구하므로 transpose
                attn_transposed = attn_out.transpose(1, 2)
                # 선형 보간법(linear interpolation)으로 업샘플링
                upsampled = F.interpolate(attn_transposed, size=target_len, mode='linear', align_corners=False)
                # 원래 형태로 다시 transpose
                upsampled_attns.append(upsampled.transpose(1, 2))
            else:
                # 길이가 같은 첫 번째 텐서는 그대로 추가
                upsampled_attns.append(attn_out)

        # 이제 모든 텐서의 길이가 같으므로 안전하게 더할 수 있음
        return sum(upsampled_attns) / len(upsampled_attns)

class PyraformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.proj = nn.Conv1d(in_ch, d, 1)
        self.pyattn = PyramidAttention(d)
        self.ff = nn.Linear(d, d)
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.proj(x).transpose(1, 2)
        x = self.pyattn(x)
        x = self.ff(x)
        return self.pool(x.transpose(1, 2))

class DilatedCNNBlock(nn.Module):
    def __init__(self, in_ch, out_ch, dilation):
        super().__init__()
        self.conv = nn.Conv1d(in_ch, out_ch, 3, padding=dilation, dilation=dilation)
        self.bn = nn.BatchNorm1d(out_ch)

    def forward(self, x):
        return F.relu(self.bn(self.conv(x)))

class DilatedCNN(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        dilations = [1, 2, 4, 8, 16]
        ch = 32
        self.layers = nn.Sequential(
            *[DilatedCNNBlock(in_ch if i==0 else ch, ch, dil) for i, dil in enumerate(dilations)]
        )
        self.out = nn.Conv1d(ch, d, 1)
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.layers(x)
        x = self.out(x)
        return self.pool(x)

class InceptionModule(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.branch1 = nn.Conv1d(in_ch, out_ch//4, 1)
        self.branch2 = nn.Conv1d(in_ch, out_ch//4, 3, padding=1)
        self.branch3 = nn.Conv1d(in_ch, out_ch//4, 5, padding=2)
        self.branch4 = nn.Sequential(nn.MaxPool1d(3, 1, 1), nn.Conv1d(in_ch, out_ch//4, 1))

    def forward(self, x):
        return torch.cat([self.branch1(x), self.branch2(x), self.branch3(x), self.branch4(x)], dim=1)

class InceptionMK(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        # 변수 이름을 'layers'로 변경
        self.layers = nn.Sequential(
            InceptionModule(in_ch, 64),
            InceptionModule(64, d)
        )
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        # 호출하는 부분도 'layers'로 변경
        x = self.layers(x)
        return self.pool(x)

# 1. Helper 클래스들 (device 처리 방식 수정됨)
class RMSNorm(nn.Module):
    def __init__(self, d, eps=1e-6):
        super().__init__()
        self.eps = eps
        self.g = nn.Parameter(torch.ones(d))

    def forward(self, x):
        return x * torch.rsqrt(x.pow(2).mean(-1, keepdim=True) + self.eps) * self.g

class DropPath(nn.Module):
    def __init__(self, drop_prob: float = 0.0):
        super().__init__()
        self.drop_prob = float(drop_prob)

    def forward(self, x):
        if self.drop_prob == 0.0 or not self.training:
            return x
        keep = 1.0 - self.drop_prob
        shape = (x.shape[0],) + (1,) * (x.ndim - 1)
        mask = x.new_empty(shape).bernoulli_(keep)
        return x * mask / keep

class MambaSSM(nn.Module):
    def __init__(self, d_model: int, d_state: int, eps: float = 1e-5):
        super().__init__()
        self.d_model = d_model
        self.d_state = d_state
        self.eps = eps
        A_init = torch.linspace(1.0, float(d_state), d_state).log()
        self.A_log = nn.Parameter(A_init)
        self.chunk_size = 32

    def forward(self, u_input, delta, b_mat, c_mat):
        B, T, D = u_input.shape
        N = self.d_state
        A = -torch.exp(self.A_log).view(1, 1, 1, N)
        chunk_size = min(self.chunk_size, T)
        num_chunks = (T + chunk_size - 1) // chunk_size
        output = torch.zeros(B, T, D, device=u_input.device, dtype=u_input.dtype)
        h = torch.zeros(B, D, N, device=u_input.device, dtype=u_input.dtype)
        for chunk_idx in range(num_chunks):
            start_idx = chunk_idx * chunk_size
            end_idx = min(start_idx + chunk_size, T)
            u_chunk = u_input[:, start_idx:end_idx, :]
            delta_chunk = delta[:, start_idx:end_idx, :]
            b_chunk = b_mat[:, start_idx:end_idx, :, :]
            c_chunk = c_mat[:, start_idx:end_idx, :, :]
            deltaA = torch.clamp(delta_chunk.unsqueeze(-1) * A, min=-10, max=10)
            exp_deltaA = torch.exp(deltaA)
            frac_exact = (exp_deltaA - 1.0) / (A + 1e-12)
            frac = torch.where(deltaA.abs() < 1e-4, delta_chunk.unsqueeze(-1), frac_exact)
            Bu = frac * (b_chunk * u_chunk.unsqueeze(-1))
            Ah = exp_deltaA
            Ah_cum = torch.cumprod(Ah, dim=1)
            h_expand = h.unsqueeze(1) * Ah_cum
            Bu_term = torch.cumsum(Bu * torch.cat([torch.ones_like(Ah[:, :1]), Ah[:, :-1].flip(1)], dim=1).flip(1), dim=1).flip(1)
            h_new = h_expand + Bu_term
            y_chunk = (h_new * c_chunk).sum(dim=-1)
            output[:, start_idx:end_idx, :] = y_chunk
            h = h_new[:, -1]
        return output

class MambaBlock(nn.Module):
    def __init__(self, d_model: int, d_state: int = 64, expansion: int = 2, groups: int = 4,
                 dropout: float = 0.1, droppath: float = 0.0, bidirectional: bool = False):
        super().__init__()
        assert (expansion * d_model) % groups == 0
        self.d_model = d_model
        self.d_state = d_state
        self.expansion = expansion
        self.d_inner = expansion * d_model
        self.groups = groups
        self.group_size = self.d_inner // groups
        self.bidirectional = bidirectional
        self.norm1 = RMSNorm(d_model)
        self.in_proj = nn.Linear(d_model, 2 * self.d_inner, bias=False)
        self.param_proj = nn.Linear(self.d_inner, self.groups * (1 + 2 * d_state), bias=False)
        self.ssm = MambaSSM(d_model=self.d_inner, d_state=d_state)
        self.out_proj = nn.Linear(self.d_inner, d_model, bias=False)
        self.drop = nn.Dropout(dropout)
        self.drop_path = DropPath(droppath) if droppath > 0.0 else nn.Identity()
        self.norm2 = RMSNorm(d_model)
        self.ffn = nn.Sequential(
            nn.Linear(d_model, 4 * d_model),
            nn.SiLU(),
            nn.Dropout(dropout),
            nn.Linear(4 * d_model, d_model)
        )

    def _expand_group_params(self, xz, params):
        B, T, _ = xz.shape
        G, N = self.groups, self.d_state
        delta_g, b_g, c_g = torch.split(params, [1, N, N], dim=-1)
        repeat = self.group_size
        delta = delta_g.repeat_interleave(repeat, dim=2).squeeze(-1)
        b_mat = b_g.unsqueeze(3).expand(B, T, G, repeat, N).reshape(B, T, -1, N)
        c_mat = c_g.unsqueeze(3).expand(B, T, G, repeat, N).reshape(B, T, -1, N)
        return delta, b_mat, c_mat

    def _run_ssm_once(self, xz, gate):
        B, T, Din = xz.shape
        raw_params = self.param_proj(xz)
        raw_params = raw_params.view(B, T, self.groups, (1 + 2 * self.d_state))
        delta, b_mat, c_mat = self._expand_group_params(xz, raw_params)
        delta = F.softplus(delta)
        y_ssm = self.ssm(xz, delta, b_mat, c_mat)
        y = y_ssm * F.silu(gate)
        return y

    def forward(self, x):
        residual = x
        z = self.norm1(x)
        xz, gate = self.in_proj(z).chunk(2, dim=-1)
        xz = F.silu(xz)
        if not self.bidirectional:
            y = self._run_ssm_once(xz, gate)
        else:
            y_f = self._run_ssm_once(xz, gate)
            y_b = self._run_ssm_once(torch.flip(xz, dims=[1]), torch.flip(gate, dims=[1]))
            y_b = torch.flip(y_b, dims=[1])
            y = 0.5 * (y_f + y_b)
        y = self.out_proj(y)
        x = residual + self.drop_path(self.drop(y))
        x = x + self.drop_path(self.drop(self.ffn(self.norm2(x))))
        return x

# 2. LiteMambaEncoder를 내부 모듈로 사용
class LiteMambaEncoder(nn.Module):
    def __init__(self, d_model: int, num_layers: int = 4, d_state: int = 64, expansion: int = 2,
                 groups: int = 4, dropout: float = 0.1, droppath: float = 0.0, use_local: bool = True,
                 bidirectional: bool = False):
        super().__init__()
        self.d_model = d_model
        self.num_layers = num_layers
        self.use_local = use_local
        self.layers = nn.ModuleList([
            MambaBlock(
                d_model=d_model,
                d_state=d_state,
                expansion=expansion,
                groups=groups,
                dropout=dropout,
                droppath=droppath * (i / (num_layers - 1)) if num_layers > 1 else droppath,
                bidirectional=bidirectional
            )
            for i in range(num_layers)
        ])
        if use_local:
            self.local_conv = nn.Sequential(
                nn.Conv1d(d_model, d_model, kernel_size=3, padding=1, groups=d_model),
                nn.BatchNorm1d(d_model),
                nn.SiLU()
            )
        self.norm = RMSNorm(d_model)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        for layer in self.layers:
            x = layer(x)
        if self.use_local:
            x_conv = x.transpose(1, 2)
            x_conv = self.local_conv(x_conv)
            x = x_conv.transpose(1, 2) + x
        x = self.norm(x)
        return x

# 3. 최종 LiteSSM 백본 클래스 (이것을 사용하면 됨)
class LiteSSM(nn.Module):
    def __init__(self, in_ch=9, d=128, num_layers=2, bidirectional=False):
        super().__init__()
        # 입력 프로젝션 레이어 (채널 수를 모델 차원에 맞춤)
        self.proj = nn.Conv1d(in_ch, d, kernel_size=1)

        # 핵심 Mamba 인코더
        self.encoder = LiteMambaEncoder(
            d_model=d,
            num_layers=num_layers,
            bidirectional=bidirectional
        )

        # 출력 풀링 레이어 (DPAHead와 호환되도록)
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        # x: (Batch, Channels, Length) e.g. [B, 9, 128]
        # 1. Projection
        x = self.proj(x) # -> [B, 128, 128]

        # 2. Mamba Encoder (입력 형식을 B, T, D로 변경)
        x = x.transpose(1, 2) # -> [B, 128, 128]
        x = self.encoder(x)
        
        # 3. Pooling (입력 형식을 B, D, T로 변경)
        x = x.transpose(1, 2) # -> [B, 128, 128]
        x = self.pool(x) # -> [B, 128, 64]
        return x

class MoELayer(nn.Module):
    def __init__(self, d, num_experts=4, topk=2):
        super().__init__()
        self.gate = nn.Linear(d, num_experts)
        self.experts = nn.ModuleList([nn.Linear(d, d) for _ in range(num_experts)])
        self.topk = topk

    def forward(self, x):
        gates = F.softmax(self.gate(x), dim=-1)
        topk_val, topk_idx = torch.topk(gates, self.topk, dim=-1)
        topk_val = topk_val / topk_val.sum(-1, keepdim=True)
        out = torch.zeros_like(x)
        for i in range(self.topk):
            mask = torch.zeros_like(gates).scatter_(-1, topk_idx[:, :, i:i+1], 1.)
            out += (mask * topk_val[:, :, i:i+1].unsqueeze(-1)) * torch.stack([exp(x) for exp in self.experts], dim=0).gather(0, topk_idx[:, :, i:i+1].unsqueeze(-1).expand(-1, -1, x.size(-1)))
        return out / self.topk

class LiteJambaBlock(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.mamba = MambaBlock(d, droppath=0.05)
        self.transformer = TransformerEncoderLayer(d)
        self.moe = MoELayer(d, 2, 1)
        self.norm = RMSNorm(d)

    def forward(self, x):
        x = self.mamba(x)
        x = self.transformer(x)
        x = self.moe(x)
        return self.norm(x)

class LiteJamba(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__()
        self.proj = nn.Conv1d(in_ch, d, 1)
        self.blocks = nn.Sequential(LiteJambaBlock(d), LiteJambaBlock(d))
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.proj(x).transpose(1, 2)
        x = self.blocks(x)
        return self.pool(x.transpose(1, 2))

class IndRNNCell(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.weight_ih = nn.Parameter(torch.randn(hidden_size, input_size))
        self.weight_hh = nn.Parameter(torch.randn(hidden_size))

    def forward(self, input, hx):
        return F.relu(torch.mm(input, self.weight_ih.t()) + hx * self.weight_hh.unsqueeze(0))

class IndRNN(nn.Module):
    def __init__(self, in_ch=9, d=128, layers=2):
        super().__init__()
        self.proj = nn.Conv1d(in_ch, d, 1)
        self.cells = nn.ModuleList([IndRNNCell(d, d) for _ in range(layers)])
        self.pool = nn.AdaptiveAvgPool1d(64)

    def forward(self, x):
        x = self.proj(x).transpose(1, 2)
        hx = [torch.zeros(x.size(0), x.size(2), device=x.device) for _ in self.cells]
        for t in range(x.size(1)):
            xt = x[:, t, :]
            for i, cell in enumerate(self.cells):
                hx[i] = cell(xt if i == 0 else hx[i-1], hx[i])
        return self.pool(hx[-1].unsqueeze(1))

class DPAHead(nn.Module):
    def __init__(self, n_classes=6, d=128, T=64, tau=0.07, smooth_lambda=1e-4):
        super().__init__()
        self.P = nn.Parameter(torch.randn(n_classes, d, T) * 0.02)
        self.tau = tau
        self.smooth_lambda = smooth_lambda

    def forward(self, h, y=None):
        B, d, T = h.shape
        sim = torch.einsum('bdt, cdt -> bct', h, self.P) / self.tau
        logits = sim.mean(dim=-1)
        attn = sim.softmax(dim=-1)
        extra_loss = 0.0
        if self.smooth_lambda > 0:
            diff = self.P[:,:,1:] - self.P[:,:,:-1]
            extra_loss = self.smooth_lambda * (diff.pow(2).mean())
        return logits, extra_loss, attn

class HARModel(nn.Module):
    def __init__(self, backbone_type, in_ch=9, d=128, n_classes=6, T=64):
        super().__init__()
        if backbone_type == "mini_resnet":
            self.backbone = MiniResNet(in_ch, d)
        elif backbone_type == "cnn_lstm":
            self.backbone = CNNLSTM(in_ch, d)
        elif backbone_type == "tcn":
            self.backbone = TCN(in_ch, d)
        elif backbone_type == "autoformer":
            self.backbone = AutoformerEncoder(in_ch, d)
        elif backbone_type == "transformer":
            self.backbone = TransformerEncoder(in_ch, d)
        elif backbone_type == "ts_transformer":
            self.backbone = TSTransformerEncoder(in_ch, d)
        elif backbone_type == "pyraformer":
            self.backbone = PyraformerEncoder(in_ch, d)
        elif backbone_type == "dilated_cnn":
            self.backbone = DilatedCNN(in_ch, d)
        elif backbone_type == "inception_mk":
            self.backbone = InceptionMK(in_ch, d)
        elif backbone_type == "lite_ssm":
            self.backbone = LiteSSM(in_ch, d)
        elif backbone_type == "lite_jamba":
            self.backbone = LiteJamba(in_ch, d)
        elif backbone_type == "indrnn":
            self.backbone = IndRNN(in_ch, d)
        else:
            raise ValueError("Unknown backbone")
        self.head = DPAHead(n_classes, d, T)

    def forward(self, x, y=None):
        h = self.backbone(x)
        logits, extra_loss, attn = self.head(h, y)
        return logits, extra_loss, attn, h

def plot_confusion_matrix(y_true, y_pred, class_names, model_name, save_path=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    if save_path:
        plt.savefig(f'{save_path}_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_tsne(features, labels, class_names, model_name, save_path=None):
    max_samples = 2000
    if len(features) > max_samples:
        indices = np.random.choice(len(features), max_samples, replace=False)
        features = features[indices]
        labels = labels[indices]

    tsne = TSNE(n_components=2, random_state=42, perplexity=30)
    features_2d = tsne.fit_transform(features)

    plt.figure(figsize=(8, 8))
    colors = plt.cm.tab10(np.linspace(0, 1, len(class_names)))

    for i, class_name in enumerate(class_names):
        mask = labels == i
        plt.scatter(features_2d[mask, 0], features_2d[mask, 1],
                   c=[colors[i]], label=class_name, alpha=0.7, s=30)

    plt.title(f't-SNE Visualization - {model_name}')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()

    if save_path:
        plt.savefig(f'{save_path}_tsne.png', dpi=300, bbox_inches='tight')
    plt.show()

def compute_metrics(model, loader, device, n_classes, class_names=None, model_name=""):
    acc = Accuracy(task="multiclass", num_classes=n_classes).to(device)
    f1 = F1Score(task="multiclass", num_classes=n_classes, average='macro').to(device)
    prec = Precision(task="multiclass", num_classes=n_classes, average='macro').to(device)
    rec = Recall(task="multiclass", num_classes=n_classes, average='macro').to(device)

    model.eval()
    all_y = []
    all_pred = []
    all_latent = []
    inf_time = 0

    with torch.no_grad():
        for batch_idx, (x, y) in enumerate(loader):
            x, y = x.to(device), y.to(device)
            start = time.time()
            logits, _, _, h = model(x)
            inf_time += (time.time() - start) * 1000 / len(x)

            pred = logits.argmax(1)
            acc.update(pred, y)
            f1.update(pred, y)
            prec.update(pred, y)
            rec.update(pred, y)

            all_y.append(y.cpu())
            all_pred.append(pred.cpu())
            all_latent.append(h.cpu().reshape(len(h), -1))

    all_y_cat = torch.cat(all_y).numpy()
    all_pred_cat = torch.cat(all_pred).numpy()
    all_latent_cat = torch.cat(all_latent).numpy()

    sil = silhouette_score(all_latent_cat, all_y_cat)
    params = sum(p.numel() for p in model.parameters())
    flops = torchprofile.profile_macs(model, torch.randn(1, 9, 128).to(device)) / 1e6

    if save_plots and class_names is not None:
        save_path = f"results_{model_name.lower()}" if model_name else None
        plot_confusion_matrix(all_y_cat, all_pred_cat, class_names, model_name, save_path)
        plot_tsne(all_latent_cat, all_y_cat, class_names, model_name, save_path)

    return {
        "accuracy": acc.compute().item(),
        "f1": f1.compute().item(),
        "precision": prec.compute().item(),
        "recall": rec.compute().item(),
        "params": params,
        "flops_m": flops,
        "silhouette": sil,
        "inf_time_ms": inf_time / len(loader)
    }

def compute_transition_metrics(model, test_ds, device, n_classes, class_names):
    model.eval()
    ratios = [(0.25, 0.75), (0.5, 0.5), (0.75, 0.25)]
    ratio_names = ["25:75", "50:50", "75:25"]

    all_results = {}
    classwise_results = {name: {} for name in ratio_names}

    for ratio_idx, (r1, r2) in enumerate(ratios):
        ratio_name = ratio_names[ratio_idx]

        kl_total = mse_total = rmse_total = mae_total = rmae_total = 0
        num_total = 0
        class_metrics = {}

        for i in range(len(test_ds) // 2):
            x1, y1 = test_ds[i]
            x2, y2 = test_ds[i + len(test_ds) // 2]
            if y1 == y2:
                continue

            with torch.no_grad():
                h1 = model.backbone(x1.unsqueeze(0).to(device))[:, :, :32]
                h2 = model.backbone(x2.unsqueeze(0).to(device))[:, :, :32]
                h_trans = torch.cat([h1, h2], dim=2)
                logits, _, attn = model.head(h_trans)
                probs = F.softmax(logits, dim=1)

                soft_label = torch.zeros(1, n_classes, device=device)
                soft_label[0, y1] = r1
                soft_label[0, y2] = r2

                kl = F.kl_div(probs.log(), soft_label, reduction='batchmean')
                mse = F.mse_loss(probs, soft_label)
                rmse = mse.sqrt()
                mae = F.l1_loss(probs, soft_label)
                rmae = mae / soft_label.mean()

                kl_total += kl.item()
                mse_total += mse.item()
                rmse_total += rmse.item()
                mae_total += mae.item()
                rmae_total += rmae.item()
                num_total += 1

                class_pair = f"{class_names[y1]}->{class_names[y2]}"
                if class_pair not in class_metrics:
                    class_metrics[class_pair] = {"kl": [], "mse": [], "rmse": [], "mae": [], "rmae": []}

                class_metrics[class_pair]["kl"].append(kl.item())
                class_metrics[class_pair]["mse"].append(mse.item())
                class_metrics[class_pair]["rmse"].append(rmse.item())
                class_metrics[class_pair]["mae"].append(mae.item())
                class_metrics[class_pair]["rmae"].append(rmae.item())

        all_results[ratio_name] = {
            "kl_div": kl_total / num_total,
            "mse": mse_total / num_total,
            "rmse": rmse_total / num_total,
            "mae": mae_total / num_total,
            "rmae": rmae_total / num_total
        }

        for class_pair, metrics in class_metrics.items():
            classwise_results[ratio_name][class_pair] = {
                "kl_div": np.mean(metrics["kl"]),
                "mse": np.mean(metrics["mse"]),
                "rmse": np.mean(metrics["rmse"]),
                "mae": np.mean(metrics["mae"]),
                "rmae": np.mean(metrics["rmae"])
            }

    return all_results, classwise_results

def train_model(model, train_loader, test_ds, epochs, bs, lr, device, class_names=None, model_name=""):
    opt = AdamW(model.parameters(), lr=lr)
    sched = CosineAnnealingLR(opt, T_max=epochs)

    # Correctly create a DataLoader for the test set
    test_loader = DataLoader(test_ds, batch_size=bs, shuffle=False)

    print(f"\nTraining {model_name} for {epochs} epochs...")

    for ep in range(epochs):
        model.train()
        epoch_loss = 0
        num_batches = 0

        for batch_idx, (x, y) in enumerate(train_loader):
            x, y = x.to(device), y.to(device)
            opt.zero_grad()
            logits, extra, _, _ = model(x, y)
            loss = F.cross_entropy(logits, y) + extra
            loss.backward()
            opt.step()

            epoch_loss += loss.item()
            num_batches += 1

        sched.step()
        avg_loss = epoch_loss / num_batches

        if (ep + 1) % 10 == 0:
            print(f"Epoch {ep+1}/{epochs}, Loss: {avg_loss:.4f}, LR: {sched.get_last_lr()[0]:.6f}")

    print(f"Training completed for {model_name}")

    # Use the new test_loader here instead of test_ds
    single_metrics = compute_metrics(model, test_loader, device, 6, class_names, model_name)

    # The transition metrics function iterates the dataset directly, so passing test_ds here is correct
    trans_metrics, classwise_trans = compute_transition_metrics(model, test_ds, device, 6, class_names)

    return single_metrics, trans_metrics, classwise_trans

def display_results_table(results_dict, table_name):
    print(f"\n{'='*60}")
    print(f"{table_name}")
    print(f"{'='*60}")

    df = pd.DataFrame(results_dict)
    df = df.round(4)
    display(df)
    return df

def main(data_root, epochs=50, bs=128, lr=1e-3):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    X_train, y_train, X_test, y_test, names = load_uci_har_raw(data_root)
    if X_train is None:
        return None

    train_ds = UCIHAR(X_train, y_train)
    test_ds = UCIHAR(X_test, y_test)
    train_loader = DataLoader(train_ds, batch_size=bs, shuffle=True)

    backbones = [
        "mini_resnet", "cnn_lstm", "tcn", "autoformer", "transformer",
        "ts_transformer", "pyraformer", "dilated_cnn", "inception_mk",
        "lite_ssm", "lite_jamba", "indrnn"
    ]

    all_results = {}
    all_single_results = {}
    all_trans_results = {}
    all_classwise_results = {}

    for bb in backbones:
        print(f"\n{'*'*80}")
        print(f"Processing Architecture: {bb.upper()}")
        print(f"{'*'*80}")

        model = HARModel(bb).to(device)
        single_metrics, trans_metrics, classwise_trans = train_model(
            model, train_loader, test_ds, epochs, bs, lr, device, names, bb
        )

        all_single_results[bb] = single_metrics
        all_trans_results[bb] = trans_metrics
        all_classwise_results[bb] = classwise_trans
        all_results[bb] = {
            "single": single_metrics,
            "transition": trans_metrics,
            "classwise": classwise_trans
        }

        print(f"\n{bb.upper()} Results Summary:")
        print("-" * 50)

        mid_rst1 = display_results_table({bb: single_metrics}, f"Unary Activity Performance - {bb.upper()}")

        df_trans_mean = pd.DataFrame(trans_metrics).T
        df_trans_mean = df_trans_mean.round(4)
        print(f"\n{'='*60}")
        print(f"Transition Activity Performance (Mean) - {bb.upper()}")
        print(f"{'='*60}")
        display(df_trans_mean)

        for ratio in ['25:75', '50:50', '75:25']:
            class_data = classwise_trans[ratio]
            if class_data:
                df_class = pd.DataFrame(class_data).T
                df_class = df_class.round(4)
                print(f"\n{'='*60}")
                print(f"Classwise Transition Performance {ratio} - {bb.upper()}")
                print(f"{'='*60}")
                display(df_class)

    print(f"\n{'='*100}")
    print("FINAL COMPREHENSIVE RESULTS")
    print(f"{'='*100}")

    rst_df = display_results_table(all_single_results, "Final Unary Activity Performance - All Models")

    trans_summary = {}
    for model in backbones:
        trans_summary[model] = {}
        for ratio in ["25:75", "50:50", "75:25"]:
            if ratio in all_trans_results[model]:
                for metric, value in all_trans_results[model][ratio].items():
                    trans_summary[model][f"{ratio}_{metric}"] = value

    trans_df = display_results_table(trans_summary, "Final Transition Performance - All Models & Ratios")

    return all_results, rst_df, trans_df

if __name__ == "__main__":
    data_root = "/workspace"
    results, final_unary_df, final_trans_df = main(data_root)
```

```python
import os
import math
import time
import random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
from torchmetrics import Accuracy, F1Score, Precision, Recall
from sklearn.metrics import silhouette_score
import torchprofile
import pandas as pd
from IPython.display import display

def load_uci_har_raw(dataset_path):
    SIGNALS = ["body_acc_x", "body_acc_y", "body_acc_z", "body_gyro_x", "body_gyro_y", "body_gyro_z", "total_acc_x", "total_acc_y", "total_acc_z"]
    X_train_list, X_test_list = [], []
    base_path_train = os.path.join(dataset_path)
    base_path_test = os.path.join(dataset_path)
    for signal in SIGNALS:
        train_file = os.path.join(base_path_train, f"{signal}_train.txt")
        test_file = os.path.join(base_path_test, f"{signal}_test.txt")
        X_train_list.append(np.loadtxt(train_file, dtype=np.float32))
        X_test_list.append(np.loadtxt(test_file, dtype=np.float32))
    X_train = np.transpose(np.stack(X_train_list, axis=-1), (0, 2, 1))
    X_test = np.transpose(np.stack(X_test_list, axis=-1), (0, 2, 1))
    y_train = np.loadtxt(os.path.join(dataset_path, 'y_train.txt'), dtype=int) - 1
    y_test = np.loadtxt(os.path.join(dataset_path, 'y_test.txt'), dtype=int) - 1
    ucihar_activity_names = ['Walking', 'Walking Upstairs', 'Walking Downstairs', 'Sitting', 'Standing', 'Laying']
    return X_train, y_train, X_test, y_test, ucihar_activity_names

class UCIHAR(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).long()
        self.mu = self.X.mean(dim=(0,2), keepdim=True)
        self.sd = self.X.std(dim=(0,2), keepdim=True).clamp_min(1e-6)
    def __len__(self):
        return len(self.y)
    def __getitem__(self, idx):
        return (self.X[idx] - self.mu.squeeze(0)) / self.sd.squeeze(0), self.y[idx]

class ResidualBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv1 = nn.Conv1d(in_ch, out_ch, 3, padding=1); self.bn1 = nn.BatchNorm1d(out_ch)
        self.conv2 = nn.Conv1d(out_ch, out_ch, 3, padding=1); self.bn2 = nn.BatchNorm1d(out_ch)
        self.shortcut = nn.Conv1d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()
    def forward(self, x):
        res = self.shortcut(x); x = F.relu(self.bn1(self.conv1(x))); x = self.bn2(self.conv2(x)); return F.relu(x + res)

class MiniResNet(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.conv = nn.Conv1d(in_ch, 32, 3, padding=1)
        self.blocks = nn.Sequential(ResidualBlock(32, 64), ResidualBlock(64, d)); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        return self.pool(self.blocks(self.conv(x)))

class CNNLSTM(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.cnn = nn.Sequential(nn.Conv1d(in_ch, 64, 5, padding=2), nn.ReLU(), nn.Conv1d(64, d, 5, padding=2), nn.ReLU())
        self.lstm = nn.LSTM(d, d, batch_first=True); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        x = self.cnn(x).transpose(1, 2); x, _ = self.lstm(x); return self.pool(x.transpose(1, 2))

class TCNBlock(nn.Module):
    def __init__(self, in_ch, out_ch, dilation):
        super().__init__(); self.conv1 = nn.Conv1d(in_ch, out_ch, 3, padding=dilation, dilation=dilation); self.bn1 = nn.BatchNorm1d(out_ch)
        self.conv2 = nn.Conv1d(out_ch, out_ch, 3, padding=dilation, dilation=dilation); self.bn2 = nn.BatchNorm1d(out_ch)
        self.shortcut = nn.Conv1d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()
    def forward(self, x):
        res = self.shortcut(x); x = F.relu(self.bn1(self.conv1(x))); x = self.bn2(self.conv2(x)); return F.relu(x + res)

class TCN(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); dilations = [1, 2, 4, 8]; ch = 64
        self.layers = nn.ModuleList([TCNBlock(in_ch if i==0 else ch, ch, dil) for i, dil in enumerate(dilations)])
        self.out = nn.Conv1d(ch, d, 1); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        for layer in self.layers: x = layer(x)
        return self.pool(self.out(x))

class SeriesDecomp(nn.Module):
    def __init__(self, kernel_size): super().__init__(); self.avg = nn.AvgPool1d(kernel_size, stride=1, padding=kernel_size//2)
    def forward(self, x): trend = self.avg(x); return x - trend, trend

class AutoformerEncoderLayer(nn.Module):
    def __init__(self, d): super().__init__(); self.decomp = SeriesDecomp(25); self.attn = nn.MultiheadAttention(d, 4, batch_first=True); self.ff = nn.Linear(d, d)
    def forward(self, x): _, _ = self.decomp(x); x = self.attn(x, x, x)[0] + x; return self.ff(x) + x

class AutoformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.proj = nn.Conv1d(in_ch, d, 1); self.layers = nn.ModuleList([AutoformerEncoderLayer(d) for _ in range(2)]); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        x = self.proj(x).transpose(1,2)
        for layer in self.layers: x = layer(x)
        return self.pool(x.transpose(1,2))

class PositionalEncoding(nn.Module):
    def __init__(self, d, max_len=128):
        super().__init__(); pe = torch.zeros(max_len, d); position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d, 2).float() * (-math.log(10000.0) / d))
        pe[:, 0::2] = torch.sin(position * div_term); pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))
    def forward(self, x): return x + self.pe[:, :x.size(1)]

class TransformerEncoderLayer(nn.Module):
    def __init__(self, d):
        super().__init__(); self.attn = nn.MultiheadAttention(d, 4, batch_first=True, dropout=0.1); self.ff = nn.Sequential(nn.Linear(d, d*4), nn.ReLU(), nn.Dropout(0.1), nn.Linear(d*4, d))
        self.norm1 = nn.LayerNorm(d); self.norm2 = nn.LayerNorm(d)
    def forward(self, x): x2 = self.norm1(x); x = x + self.attn(x2, x2, x2)[0]; x2 = self.norm2(x); return x + self.ff(x2)

class TransformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.proj = nn.Linear(in_ch, d); self.pe = PositionalEncoding(d)
        self.layers = nn.ModuleList([TransformerEncoderLayer(d) for _ in range(2)]); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        x = self.proj(x.transpose(1, 2)); x = self.pe(x)
        for layer in self.layers: x = layer(x)
        return self.pool(x.transpose(1, 2))

class TSTransformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.proj = nn.Conv1d(in_ch, d, 1); self.pe = PositionalEncoding(d, 128)
        self.layers = nn.ModuleList([TransformerEncoderLayer(d) for _ in range(2)]); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        x = self.proj(x).transpose(1, 2); x = self.pe(x)
        for layer in self.layers: x = layer(x)
        return self.pool(x.transpose(1, 2))

class PyramidAttention(nn.Module):
    def __init__(self, d): super().__init__(); self.attn = nn.MultiheadAttention(d, 4, batch_first=True)
    def forward(self, x):
        levels = [x[:, ::2**i] for i in range(3)]; attns = [self.attn(l, l, l)[0] for l in levels]; target_len = attns[0].size(1); upsampled_attns = []
        for attn_out in attns:
            if attn_out.size(1) < target_len: upsampled_attns.append(F.interpolate(attn_out.transpose(1, 2), size=target_len, mode='linear', align_corners=False).transpose(1, 2))
            else: upsampled_attns.append(attn_out)
        return sum(upsampled_attns) / len(upsampled_attns)

class PyraformerEncoder(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.proj = nn.Conv1d(in_ch, d, 1); self.pyattn = PyramidAttention(d); self.ff = nn.Linear(d, d); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x): x = self.proj(x).transpose(1, 2); x = self.pyattn(x); x = self.ff(x); return self.pool(x.transpose(1, 2))

class DilatedCNN(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); ch = 32; dilations = [1, 2, 4, 8, 16]
        self.layers = nn.Sequential(*[nn.Sequential(nn.Conv1d(in_ch if i==0 else ch, ch, 3, padding=dil, dilation=dil), nn.BatchNorm1d(ch), nn.ReLU()) for i, dil in enumerate(dilations)])
        self.out = nn.Conv1d(ch, d, 1); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x): return self.pool(self.out(self.layers(x)))

class InceptionModule(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__(); co = out_ch // 4; self.b1 = nn.Conv1d(in_ch, co, 1); self.b2 = nn.Conv1d(in_ch, co, 3, padding=1)
        self.b3 = nn.Conv1d(in_ch, co, 5, padding=2); self.b4 = nn.Sequential(nn.MaxPool1d(3, 1, 1), nn.Conv1d(in_ch, out_ch - 3*co, 1))
    def forward(self, x): return torch.cat([self.b1(x), self.b2(x), self.b3(x), self.b4(x)], dim=1)

class InceptionMK(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.layers = nn.Sequential(InceptionModule(in_ch, 64), InceptionModule(64, d)); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x): return self.pool(self.layers(x))

class RMSNorm(nn.Module):
    def __init__(self, d, eps=1e-6):
        super().__init__(); self.eps = eps; self.g = nn.Parameter(torch.ones(d))
    def forward(self, x):
        return x * torch.rsqrt(x.pow(2).mean(-1, keepdim=True) + self.eps) * self.g

class DropPath(nn.Module):
    def __init__(self, drop_prob: float = 0.0):
        super().__init__(); self.drop_prob = float(drop_prob)
    def forward(self, x):
        if self.drop_prob == 0.0 or not self.training: return x
        keep = 1.0 - self.drop_prob; shape = (x.shape[0],) + (1,) * (x.ndim - 1)
        mask = x.new_empty(shape).bernoulli_(keep); return x * mask / keep

class MambaSSM(nn.Module):
    def __init__(self, d_model: int, d_state: int, eps: float = 1e-5):
        super().__init__(); self.d_model, self.d_state, self.eps = d_model, d_state, eps
        self.A_log = nn.Parameter(torch.linspace(1.0, float(d_state), d_state).log())
        self.chunk_size = 32
    def forward(self, u_input, delta, b_mat, c_mat):
        B, T, D = u_input.shape; N = self.d_state; A = -torch.exp(self.A_log).view(1, 1, 1, N)
        chunk_size = min(self.chunk_size, T); num_chunks = (T + chunk_size - 1) // chunk_size
        output = torch.zeros(B, T, D, device=u_input.device, dtype=u_input.dtype)
        h = torch.zeros(B, D, N, device=u_input.device, dtype=u_input.dtype)
        for chunk_idx in range(num_chunks):
            start_idx, end_idx = chunk_idx * chunk_size, min((chunk_idx + 1) * chunk_size, T)
            u_chunk, delta_chunk = u_input[:, start_idx:end_idx, :], delta[:, start_idx:end_idx, :]
            b_chunk, c_chunk = b_mat[:, start_idx:end_idx, :, :], c_mat[:, start_idx:end_idx, :, :]
            deltaA = torch.clamp(delta_chunk.unsqueeze(-1) * A, min=-10, max=10); exp_deltaA = torch.exp(deltaA)
            frac_exact = (exp_deltaA - 1.0) / (A + 1e-12); frac = torch.where(deltaA.abs() < 1e-4, delta_chunk.unsqueeze(-1), frac_exact)
            Bu = frac * (b_chunk * u_chunk.unsqueeze(-1)); Ah = exp_deltaA; Ah_cum = torch.cumprod(Ah, dim=1)
            h_expand = h.unsqueeze(1) * Ah_cum
            Bu_term = torch.cumsum(Bu * torch.cat([torch.ones_like(Ah[:, :1]), Ah[:, :-1].flip(1)], dim=1).flip(1), dim=1).flip(1)
            h_new = h_expand + Bu_term; y_chunk = (h_new * c_chunk).sum(dim=-1)
            output[:, start_idx:end_idx, :] = y_chunk; h = h_new[:, -1]
        return output

class MambaBlock(nn.Module):
    def __init__(self, d_model: int, d_state: int = 64, expansion: int = 2, groups: int = 4, dropout: float = 0.1, droppath: float = 0.0, bidirectional: bool = False):
        super().__init__(); self.d_model, self.d_state, self.expansion, self.groups, self.bidirectional = d_model, d_state, expansion, groups, bidirectional
        self.d_inner = expansion * d_model; self.group_size = self.d_inner // groups
        self.norm1 = RMSNorm(d_model); self.in_proj = nn.Linear(d_model, 2 * self.d_inner, bias=False)
        self.param_proj = nn.Linear(self.d_inner, self.groups * (1 + 2 * d_state), bias=False); self.ssm = MambaSSM(d_model=self.d_inner, d_state=d_state)
        self.out_proj = nn.Linear(self.d_inner, d_model, bias=False); self.drop = nn.Dropout(dropout)
        self.drop_path = DropPath(droppath) if droppath > 0.0 else nn.Identity()
        self.norm2 = RMSNorm(d_model); self.ffn = nn.Sequential(nn.Linear(d_model, 4*d_model), nn.SiLU(), nn.Dropout(dropout), nn.Linear(4*d_model, d_model))
    def _expand_group_params(self, xz, params):
        B, T, _ = xz.shape; G, N = self.groups, self.d_state; delta_g, b_g, c_g = torch.split(params, [1, N, N], dim=-1)
        repeat = self.group_size; delta = delta_g.repeat_interleave(repeat, dim=2).squeeze(-1)
        b_mat = b_g.unsqueeze(3).expand(B, T, G, repeat, N).reshape(B, T, -1, N)
        c_mat = c_g.unsqueeze(3).expand(B, T, G, repeat, N).reshape(B, T, -1, N)
        return delta, b_mat, c_mat
    def _run_ssm_once(self, xz, gate):
        B, T, Din = xz.shape; raw_params = self.param_proj(xz); raw_params = raw_params.view(B, T, self.groups, (1 + 2 * self.d_state))
        delta, b_mat, c_mat = self._expand_group_params(xz, raw_params); delta = F.softplus(delta)
        y_ssm = self.ssm(xz, delta, b_mat, c_mat); y = y_ssm * F.silu(gate); return y
    def forward(self, x):
        residual = x; z = self.norm1(x); xz, gate = self.in_proj(z).chunk(2, dim=-1); xz = F.silu(xz)
        if not self.bidirectional: y = self._run_ssm_once(xz, gate)
        else:
            y_f = self._run_ssm_once(xz, gate); y_b = self._run_ssm_once(torch.flip(xz, dims=[1]), torch.flip(gate, dims=[1]))
            y = 0.5 * (y_f + torch.flip(y_b, dims=[1]))
        y = self.out_proj(y); x = residual + self.drop_path(self.drop(y)); x = x + self.drop_path(self.drop(self.ffn(self.norm2(x)))); return x

class LiteMambaEncoder(nn.Module):
    def __init__(self, d_model: int, num_layers: int = 4, d_state: int = 64, expansion: int = 2, groups: int = 4, dropout: float = 0.1, droppath: float = 0.0, use_local: bool = True, bidirectional: bool = False):
        super().__init__(); self.d_model, self.num_layers, self.use_local = d_model, num_layers, use_local
        self.layers = nn.ModuleList([MambaBlock(d_model=d_model, d_state=d_state, expansion=expansion, groups=groups, dropout=dropout, droppath=droppath*(i/(num_layers-1)) if num_layers>1 else droppath, bidirectional=bidirectional) for i in range(num_layers)])
        if use_local: self.local_conv = nn.Sequential(nn.Conv1d(d_model, d_model, kernel_size=3, padding=1, groups=d_model), nn.BatchNorm1d(d_model), nn.SiLU())
        self.norm = RMSNorm(d_model)
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        for layer in self.layers: x = layer(x)
        if self.use_local: x_conv = self.local_conv(x.transpose(1, 2)); x = x_conv.transpose(1, 2) + x
        return self.norm(x)

class LiteSSM(nn.Module):
    def __init__(self, in_ch=9, d=128, num_layers=2, bidirectional=False):
        super().__init__(); self.proj = nn.Conv1d(in_ch, d, kernel_size=1)
        self.encoder = LiteMambaEncoder(d_model=d, num_layers=num_layers, bidirectional=bidirectional, d_state=16)
        self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        x = self.proj(x).transpose(1, 2); x = self.encoder(x); return self.pool(x.transpose(1, 2))

class MoELayer(nn.Module):
    def __init__(self, d, num_experts=4, topk=2):
        super().__init__(); self.gate = nn.Linear(d, num_experts); self.experts = nn.ModuleList([nn.Linear(d, d) for _ in range(num_experts)]); self.topk = topk
    def forward(self, x):
        gates = F.softmax(self.gate(x), dim=-1); topk_val, topk_idx = torch.topk(gates, self.topk, dim=-1)
        topk_val /= (topk_val.sum(dim=-1, keepdim=True) + 1e-6); y = torch.zeros_like(x)
        for i in range(self.topk):
            expert_outputs = torch.stack([expert(x) for expert in self.experts], dim=2)
            idx_expanded = topk_idx[:, :, i:i+1].unsqueeze(-1).expand(-1, -1, -1, x.shape[-1])
            selected_outputs = torch.gather(expert_outputs, 2, idx_expanded).squeeze(2); y += selected_outputs * topk_val[:, :, i:i+1]
        return y

class LiteJambaBlock(nn.Module):
    def __init__(self, d): super().__init__(); self.mamba = MambaBlock(d, d_state=16); self.transformer = TransformerEncoderLayer(d); self.moe = MoELayer(d, 4, 1); self.norm = RMSNorm(d)
    def forward(self, x): x = self.mamba(x); x = self.transformer(x); x = self.moe(x); return self.norm(x)

class LiteJamba(nn.Module):
    def __init__(self, in_ch=9, d=128):
        super().__init__(); self.proj = nn.Conv1d(in_ch, d, 1); self.blocks = nn.Sequential(LiteJambaBlock(d), LiteJambaBlock(d)); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x): x = self.proj(x).transpose(1, 2); x = self.blocks(x); return self.pool(x.transpose(1, 2))

class IndRNNCell(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__(); self.weight_ih = nn.Parameter(torch.randn(hidden_size, input_size)); self.weight_hh = nn.Parameter(torch.randn(hidden_size))
    def forward(self, x, hx): return F.relu(torch.matmul(x, self.weight_ih.t()) + hx * self.weight_hh)

class IndRNN(nn.Module):
    def __init__(self, in_ch=9, d=128, num_layers=2):
        super().__init__(); self.proj = nn.Conv1d(in_ch, d, 1); self.rnn_layers = nn.ModuleList([IndRNNCell(d, d) for _ in range(num_layers)]); self.pool = nn.AdaptiveAvgPool1d(64)
    def forward(self, x):
        x = self.proj(x).transpose(1, 2); B, T, D = x.shape; hx = torch.zeros(B, D, device=x.device)
        outputs = []
        for t in range(T):
            h_in = x[:, t, :]; h_out = h_in
            for rnn_layer in self.rnn_layers: h_out = rnn_layer(h_out, hx)
            hx = h_out; outputs.append(hx)
        return self.pool(torch.stack(outputs, dim=1).transpose(1, 2))

class DPAHead(nn.Module):
    def __init__(self, n_classes=6, d=128, T=64, tau=0.07, smooth_lambda=1e-4):
        super().__init__(); self.P = nn.Parameter(torch.randn(n_classes, d, T) * 0.02); self.tau = tau; self.smooth_lambda = smooth_lambda
    def forward(self, h, y=None):
        sim = torch.einsum('bdt, cdt -> bct', h, self.P) / self.tau; logits = sim.mean(dim=-1); extra_loss = 0.0
        if self.smooth_lambda > 0 and self.training: diff = self.P[:,:,1:] - self.P[:,:,:-1]; extra_loss = self.smooth_lambda * (diff.pow(2).mean())
        return logits, extra_loss, None

class HARModel(nn.Module):
    def __init__(self, backbone_type, head_type='dpa', in_ch=9, d=128, n_classes=6, T_pooled=64):
        super().__init__(); self.backbone_type, self.head_type = backbone_type, head_type
        if backbone_type == "mini_resnet": self.backbone = MiniResNet(in_ch, d)
        elif backbone_type == "cnn_lstm": self.backbone = CNNLSTM(in_ch, d)
        elif backbone_type == "tcn": self.backbone = TCN(in_ch, d)
        elif backbone_type == "autoformer": self.backbone = AutoformerEncoder(in_ch, d)
        elif backbone_type == "transformer": self.backbone = TransformerEncoder(in_ch, d)
        elif backbone_type == "ts_transformer": self.backbone = TSTransformerEncoder(in_ch, d)
        elif backbone_type == "pyraformer": self.backbone = PyraformerEncoder(in_ch, d)
        elif backbone_type == "dilated_cnn": self.backbone = DilatedCNN(in_ch, d)
        elif backbone_type == "inception_mk": self.backbone = InceptionMK(in_ch, d)
        elif backbone_type == "lite_ssm": self.backbone = LiteSSM(in_ch, d)
        elif backbone_type == "lite_jamba": self.backbone = LiteJamba(in_ch, d)
        elif backbone_type == "indrnn": self.backbone = IndRNN(in_ch, d)
        else: raise ValueError(f"Unsupported backbone: {backbone_type}")
        if head_type == 'dpa': self.head = DPAHead(n_classes, d, T_pooled)
        elif head_type == 'linear': self.head = nn.Sequential(nn.Flatten(), nn.Linear(d * T_pooled, n_classes))
        else: raise ValueError(f"Unsupported head_type: {head_type}")
    def forward(self, x, y=None):
        h = self.backbone(x)
        if self.head_type == 'dpa': logits, extra_loss, attn = self.head(h, y); return logits, extra_loss, attn, h
        else: logits = self.head(h); return logits, 0.0, None, h

def compute_metrics(model, loader, device, n_classes):
    metric_fns = {'accuracy': Accuracy(task="multiclass", num_classes=n_classes).to(device),'f1': F1Score(task="multiclass", num_classes=n_classes, average='macro').to(device),'precision': Precision(task="multiclass", num_classes=n_classes, average='macro').to(device),'recall': Recall(task="multiclass", num_classes=n_classes, average='macro').to(device)}
    model.eval(); all_latent, all_y, inf_time = [], [], 0
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            if device.type == 'cuda': torch.cuda.synchronize()
            start = time.time()
            logits, _, _, h = model(x)
            if device.type == 'cuda': torch.cuda.synchronize()
            end = time.time()
            inf_time += (end - start) * 1000 / len(x)
            pred = logits.argmax(1)
            for fn in metric_fns.values(): fn.update(pred, y)
            all_latent.append(h.cpu().reshape(len(h), -1)); all_y.append(y.cpu())
    results = {name: fn.compute().item() for name, fn in metric_fns.items()}
    all_y_cat = torch.cat(all_y).numpy(); all_latent_cat = torch.cat(all_latent).numpy()
    results["silhouette"] = silhouette_score(all_latent_cat, all_y_cat)
    results["inf_time(ms)"] = inf_time / len(loader)
    return results

def compute_transition_metrics(model, test_ds, device, n_classes):
    model.eval(); ratios = [(0.25, 0.75), (0.5, 0.5), (0.75, 0.25)]; all_results = {}
    for r1, r2 in ratios:
        metrics = {'kl_div': [], 'mse': [], 'rmse': [], 'mae': [], 'rmae': []}
        for i in range(len(test_ds) // 2):
            x1, y1 = test_ds[i]; x2, y2 = test_ds[i + len(test_ds) // 2]
            if y1 == y2: continue
            with torch.no_grad():
                h1 = model.backbone(x1.unsqueeze(0).to(device)); h2 = model.backbone(x2.unsqueeze(0).to(device))
                h1_pooled = F.adaptive_avg_pool1d(h1, 32); h2_pooled = F.adaptive_avg_pool1d(h2, 32)
                h_trans = torch.cat([h1_pooled, h2_pooled], dim=-1)
                if model.head_type == 'dpa': logits, _, _ = model.head(h_trans)
                else: logits = model.head(h_trans)
                probs = F.softmax(logits, dim=1)
                soft_label = torch.zeros(1, n_classes, device=device); soft_label[0, y1] = r1; soft_label[0, y2] = r2
                metrics['kl_div'].append(F.kl_div(probs.log(), soft_label, reduction='batchmean').item())
                mse = F.mse_loss(probs, soft_label).item(); metrics['mse'].append(mse); metrics['rmse'].append(math.sqrt(mse))
                mae = F.l1_loss(probs, soft_label).item(); metrics['mae'].append(mae); metrics['rmae'].append(mae / soft_label.mean().item())
        if metrics['kl_div']:
            ratio_name = f"{int(r1*100)}:{int(r2*100)}"; all_results[ratio_name] = {k: np.mean(v) for k, v in metrics.items()}
    return all_results

def train_model(model, train_loader, test_loader, test_ds, epochs, lr, device, n_classes, model_name=""):
    opt = AdamW(model.parameters(), lr=lr); sched = CosineAnnealingLR(opt, T_max=epochs)
    print(f"Training {model_name} for {epochs} epochs...")
    for ep in range(epochs):
        model.train(); total_loss = 0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device); opt.zero_grad()
            logits, extra, _, _ = model(x, y)
            loss = F.cross_entropy(logits, y) + extra; loss.backward(); opt.step(); total_loss += loss.item()
        sched.step()
        print(f"Epoch {ep+1}/{epochs}, Loss: {total_loss/len(train_loader):.4f}")
    print(f"Training completed for {model_name}")
    single_metrics = compute_metrics(model, test_loader, device, n_classes)
    trans_metrics = compute_transition_metrics(model, test_ds, device, n_classes)
    return single_metrics, trans_metrics

def display_results_table(results_dict, table_name):
    print(f"\n{'='*120}\n{table_name}\n{'='*120}")
    df = pd.DataFrame(results_dict).T
    pd.set_option('display.float_format', '{:.4f}'.format); display(df)
    return df

def main(data_root, epochs=30, bs=128, lr=1e-3):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu"); print(f"Using device: {device}")
    X_train, y_train, X_test, y_test, names = load_uci_har_raw(data_root)
    train_ds = UCIHAR(X_train, y_train); test_ds = UCIHAR(X_test, y_test)
    train_loader = DataLoader(train_ds, batch_size=bs, shuffle=True, num_workers=2, pin_memory=True)
    test_loader = DataLoader(test_ds, batch_size=bs*2, shuffle=False, num_workers=2, pin_memory=True)
    
    backbone_list = ["lite_ssm", "lite_jamba", "mini_resnet", "cnn_lstm", "tcn",
                     "autoformer", "transformer", "ts_transformer", 
                     "pyraformer", "dilated_cnn", "inception_mk"]
    experiments = [{'name': f"{bb.upper()} with DPA", 'backbone': bb, 'head': 'dpa'} for bb in backbone_list] + \
                  [{'name': f"{bb.upper()} with Linear", 'backbone': bb, 'head': 'linear'} for bb in backbone_list]

    all_single_results, all_trans_results = {}, {}
    for exp in experiments:
        model_name = exp['name']; print(f"\n{'*'*80}\nProcessing: {model_name}\n{'*'*80}")
        model = HARModel(backbone_type=exp['backbone'], head_type=exp['head'], n_classes=len(names)).to(device)
        
        single, trans = train_model(model, train_loader, test_loader, test_ds, epochs, lr, device, len(names), model_name)
        
        params = sum(p.numel() for p in model.parameters()); flops_macs = torchprofile.profile_macs(model, torch.randn(1, 9, 128).to(device))
        single['params(M)'] = params / 1e6; single['flops(M)'] = flops_macs / 1e6
        
        all_single_results[model_name] = single
        all_trans_results[model_name] = trans
        
    final_unary_df = display_results_table(all_single_results, "Final Unary Activity Performance (DPA vs Linear)")
    trans_summary = {model: {f"{ratio}_{metric}": value for ratio, metrics in data.items() for metric, value in metrics.items()} for model, data in all_trans_results.items() if data}
    if trans_summary:
        final_trans_df = display_results_table(trans_summary, "Final Transition Performance (DPA vs Linear)")

if __name__ == "__main__":
    data_root = "/workspace" 
    main(data_root, epochs=30, bs=256 , lr=1e-3)
```
